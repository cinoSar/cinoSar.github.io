## 1. From Hypothesis Testing to Estimation

### The Limitation of Binary Tests

Hypothesis tests answer a yes/no question: "Do we reject H0?"

They give us:
- **Reject** or **Fail to Reject** the null hypothesis
- A **p-value** indicating evidence against H0

But they don't tell us: *What range of values is plausible for the parameter?*

**Example:** Testing H0: mu=25000 (graduate salary) gives us p-value, but what if the true mean is 25200? Or 26500?

### The Confidence Interval Approach

A **confidence interval (CI)** is an interval [L, U] constructed from sample data such that:
- It contains a range of plausible values for the population parameter
- The interval construction method has a specified confidence level (e.g., 95%)

### Mathematical Connection to Tests

**Key Insight:** A (1-alpha)*100% confidence interval for mu is the set of all values mu0 that would NOT be rejected by a two-sided test of H0: mu=mu0 at significance level alpha.

In other words:
- If mu0 falls inside the 95% CI, then a two-sided t-test of H0: mu=mu0 would NOT reject at alpha=0.05
- If mu0 falls outside the 95% CI, then the test WOULD reject at alpha=0.05

## 2. Confidence Interval for the Mean (sigma unknown)

### The Formula

When the population standard deviation sigma is unknown (the typical case), we use the **t-distribution**:

X-bar ± t[n-1, alpha/2] * s/sqrt(n)

Where:
- **X-bar** = sample mean
- **t[n-1, alpha/2]** = critical value from t-distribution with (n-1) degrees of freedom
- **s** = sample standard deviation
- **s/sqrt(n)** = **Standard Error (SE)** — the standard deviation of X-bar
- **n** = sample size

### Breaking Down the Components

**Standard Error:** SE = s/sqrt(n)

This measures the precision of the sample mean estimate:
- Larger sample size → smaller SE → more precise estimate
- More variable population → larger SE → less precise estimate
- SE is proportional to 1/sqrt(n) (not 1/n)

**Margin of Error:** ME = t[n-1, alpha/2] * SE

The CI can be written as: X-bar ± ME

### Worked Example: Graduate Salaries

A sample of 64 recent graduates has:
- Sample mean: X-bar = 26,200
- Sample SD: s = 4,800
- We want a 95% CI

**Step 1: Calculate SE**

SE = s/sqrt(n) = 4,800/sqrt(64) = 4,800/8 = 600

**Step 2: Find the critical value**

For 95% CI: alpha = 0.05, so alpha/2 = 0.025
Degrees of freedom: n - 1 = 63
From t-table or calculator: t[63, 0.025] ≈ 2.00

**Step 3: Calculate ME and CI**

ME = 2.00 * 600 = 1,200
CI = 26,200 ± 1,200 = [25,000, 27,400]

**Interpretation:** We are 95% confident that the population mean salary lies between 25,000 and 27,400.

### Common Mistake: SE denominator

**WRONG:** SE = s * sqrt(n) (multiply by square root)
**RIGHT:** SE = s/sqrt(n) (divide by square root)

The reason: as n increases, individual observations become less important—the average of more numbers is more stable, not less.

```python
# Worked Example: Graduate Salaries
n = 64
x_bar = 26200
s = 4800
confidence = 0.95

# Step 1: Standard Error
se = s / np.sqrt(n)
print(f"Sample size: {n}")
print(f"Sample mean: ${x_bar:,.0f}")
print(f"Sample SD: ${s:,.0f}")
print(f"Standard Error: ${se:,.0f}")

# Step 2: Critical value
alpha = 1 - confidence
df = n - 1
t_crit = stats.t.ppf(1 - alpha/2, df)
print(f"\nDegrees of freedom: {df}")
print(f"alpha/2: {alpha/2}")
print(f"t-critical value: {t_crit:.4f}")

# Step 3: CI
me = t_crit * se
ci_lower = x_bar - me
ci_upper = x_bar + me

print(f"\nMargin of Error: ${me:,.0f}")
print(f"95% Confidence Interval: [${ci_lower:,.0f}, ${ci_upper:,.0f}]")
```

## 3. Understanding t[n-1, α/2]### Breaking Down the Subscripts**n-1 = Degrees of Freedom (df)**- When we estimate sigma from the sample (using s), we lose one degree of freedom- Why? Because s is calculated from the mean, which constrains one observation- df = n - 1 because we estimate one parameter (mu) from the sample**α/2 = Upper Tail Probability**- The error probability alpha is split equally between both tails- For 95% CI: alpha = 0.05, so we look for P(T > t) = 0.025

## 4. Correct Interpretation of Confidence Intervals**CORRECT:**If we repeated this study many times, 95% of the CIs would contain the true parameter.The confidence is in the *method* of construction, not in any single interval.**WRONG:**There is a 95% probability that mu lies in this interval.Once calculated, an interval either contains mu (probability=1) or doesn't (probability=0).The 95% refers to the long-run behavior of the procedure.

## 5. CI and Hypothesis Test Duality**Theorem:** A value mu0 falls inside the (1-alpha)*100% CI if and only if we would NOT reject H0: mu=mu0 in a two-sided test at significance level alpha.In other words:- **mu0 inside 95% CI** <=> **Cannot reject H0: mu=mu0 at alpha=0.05**- **mu0 outside 95% CI** <=> **Would reject H0: mu=mu0 at alpha=0.05**

## 6. Effect of Sample SizeSE is inversely proportional to sqrt(n):- Doubling n reduces SE by sqrt(2) ≈ 1.41- To halve SE, we need to quadruple n**Sample Size Formula:**n = (z[alpha/2] * s / E)^2Where E is the desired margin of error.

## 7. Effect of Confidence LevelHigher confidence level → Wider interval (precision-confidence trade-off)For n=64, X-bar=26200, s=4800:- 90% CI: [25,200, 27,200] (ME = 1,000)- 95% CI: [25,000, 27,400] (ME = 1,200)- 99% CI: [24,604, 27,796] (ME = 1,596)

## 8. Confidence Interval for Proportionsp-hat ± z[alpha/2] * sqrt(p-hat(1-p-hat)/n)**Validity Conditions:**- np-hat >= 5- n(1-p-hat) >= 5**Key Insight:** SE is maximized when p-hat = 0.5 (maximum uncertainty).

## 9. Confidence Interval for Difference of MeansFor independent samples:(X-bar1 - X-bar2) ± t* * sqrt(s1^2/n1 + s2^2/n2)**Interpretation:**- If CI contains 0: difference is NOT statistically significant- If CI entirely positive: Group 1 mean > Group 2 mean- If CI entirely negative: Group 1 mean < Group 2 mean

## Key Takeaways

1. CIs extend hypothesis testing by providing a range of plausible values
2. Formula for mean: X-bar ± t[n-1,a/2] * s/sqrt(n)
3. CI-test duality: mu0 inside 95% CI iff cannot reject H0: mu=mu0 at a=0.05
4. Correct interpretation: confidence in procedure, not in single interval
5. Sample size: halving SE requires quadrupling n
6. Precision-confidence trade-off: higher confidence → wider intervals
7. Proportions: use p-hat ± z[a/2]*sqrt(p-hat(1-p-hat)/n) with validity checks
8. Difference of means: CI tests whether groups differ significantly
